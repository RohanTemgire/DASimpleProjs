```python
import pandas as pd
```

### 1) Find all the null values from the dataset and replace it with the mean of their column

```python
data.isnull().sum()
```

```python
# dropping from make,model,type, origin cause cant make mean and replace them
data.dropna(subset = ['Make','Model','Type','Origin','DriveTrain'],inplace = True)
```

```python
data.isnull().sum()
```

```python
#  now replacing cylinders with their null values
data['Cylinders'].fillna(data['Cylinders'].mean(), inplace = True)
```

```python
data.isnull().sum()
```

### 2) Check what are the different types of make in the dataset and count of each

```python
data['Make'].value_counts()
```

### 3) Show all the records where the origin is Aisa or Europe

```python
data[(data.Origin == 'Aisa') | (data.Origin == 'Europe')]
```

### 4) remove all the records where the weight is above 4000

```python
data.head(2)
```

```python
# ~ sign is used to remove the records

data[~(data['Weight'] > 4000)]
```

```python
data.shape
```

### 5) Increase all the values of the column 'MPG_City' by 3

```python
data['MPG_City'] = data['MPG_City'].apply(lambda x: x+3)
```
